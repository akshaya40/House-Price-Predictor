# **House Price Predictor**

## **Loading the required dataset**

```python
from google.colab import files
uploaded = files.upload()

```

```python
import pandas as pd
df=pd.read_csv("HousePrices (1).csv")
df.head()
```

## **Splitting of data**

```python
y=df['Price']
X = df.drop(['Price', 'Date', 'id'], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train

```

```python
X_test
```

# **Model Building**

## **Linear Regression**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

### **Making a Prediction**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
arr=y_lr_train_pred
series = pd.Series(arr)
print(series)
```

```python
arr1=y_lr_test_pred
series = pd.Series(arr1)
print(series)
```

# **Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score,accuracy_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)


lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

# **Random Forest**

## **Training the Model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### **Making the Prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **Evaluating Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)


rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Comparing Both Models**

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```

### **Key Points:**
1. Linear Regression: Better fit and accuracy with lower errors and higher explained variance.
2. Random Forest: Higher errors and lower explained variance, making it less effective in this case.


Overall, Linear Regression is the more effective model for this dataset.

# **Data Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5, 5))
plt.scatter(x=y_train, y=y_lr_train_pred, c="#7CAE00", alpha=0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Predicted House Prices')
plt.xlabel('Actual House Prices')
plt.title('Actual vs Predicted House Prices')

plt.show()
```

The scatter plot illustrates the relationship between actual and predicted house prices. Each dot represents a single house. The red line represents the ideal prediction (where actual and predicted prices match). The closer the dots are to the red line, the more accurate the predictions.
